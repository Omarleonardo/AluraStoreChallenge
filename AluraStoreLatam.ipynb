### Importación de datos



```python
from matplotlib import pyplot as plt
```

```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda4.head()
```

#1. Análisis de facturación



```python
ventas_tienda1 = tienda['Precio'].sum()
ventas_tienda2 = tienda2['Precio'].sum()
ventas_tienda3 = tienda3['Precio'].sum()
ventas_tienda4 = tienda4['Precio'].sum()
```

```python
print("ANÁLISIS DE INGRESOS TOTALES POR TIENDA \n")
print(f"Ingreso Total Tienda 1: ${ventas_tienda1:,.2f}")
print(f"Ingreso Total Tienda 2: ${ventas_tienda2:,.2f}")
print(f"Ingreso Total Tienda 3: ${ventas_tienda3:,.2f}")
print(f"Ingreso Total Tienda 4: ${ventas_tienda4:,.2f}")
```

```python
ventas_por_tienda = {
    "Tienda 1" : ventas_tienda1,
    "Tienda 2" : ventas_tienda2,
    "Tienda 3" : ventas_tienda3,
    "Tienda 4" : ventas_tienda4
    }
tienda_mayor_ingreso = max(ventas_por_tienda, key=ventas_por_tienda.get)
print(f"Tienda con mayor ingreso: {tienda_mayor_ingreso}")
print(f"Ingreso: ${ventas_por_tienda[tienda_mayor_ingreso]:,.2f}")
```

# 2. Ventas por categoría

```python
def ventas_por_categoria(tienda):
  cantidad_vendida = 0
  ventas_categoria = tienda['Categoría del Producto'].value_counts()
  for categoria, cantidad in ventas_categoria.items():
    cantidad_vendida += cantidad
    print(f"  {categoria}: {cantidad} productos")

  categoria_mas_popular = ventas_categoria.index[0]
  max_ventas = ventas_categoria.iloc[0]
  return f"Categoría más popular: {categoria_mas_popular} ({max_ventas} productos)", f"Ventas totales: {cantidad_vendida}"
```

```python
tiendas = {
    'Tienda 1': tienda,
    'Tienda 2': tienda2,
    'Tienda 3': tienda3,
    'Tienda 4': tienda4
}
for nombre_tienda, tienda in tiendas.items():
  print(f"{nombre_tienda}")
  print("Cantidad de productos vendidos por categoría:")

  print(f"{ventas_por_categoria(tienda)} \n")
```

# 3. Calificación promedio de la tienda


```python
valoracion_tienda1 = tienda['Calificación'].mean()
valoracion_tienda2 = tienda2['Calificación'].mean()
valoracion_tienda3 = tienda3['Calificación'].mean()
valoracion_tienda4 = tienda4['Calificación'].mean()
```

```python
print("Valoración media por tienda \n")
print(f"Valoración Tienda 1: {valoracion_tienda1:.2f}")
print(f"Valoración Tienda 2: {valoracion_tienda2:.2f}")
print(f"Valoración Tienda 3: {valoracion_tienda3:.2f}")
print(f"Valoración Tienda 4: {valoracion_tienda4:.2f}")
```

```python
valoracion_por_tienda = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Valoración por tienda': [valoracion_tienda1, valoracion_tienda2, valoracion_tienda3, valoracion_tienda4]
})
valoracion_ordenada = valoracion_por_tienda.sort_values('Valoración por tienda', ascending=True)
print("\nTiendas ordenadas por valoración (menor a mayor):")
print(valoracion_ordenada)
```

```python
plt.figure(figsize=(10, 6))
plt.barh(valoracion_ordenada['Tienda'],
        valoracion_ordenada['Valoración por tienda'],
        color=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99'])

plt.title('Valoración de los cliente por tienda', fontsize=14, fontweight='bold')
plt.xlabel('Calificación', fontsize=12)
plt.ylabel('', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python
def productos_mas_y_menos_vendidos(tienda):
  productos_mas_vendidos = tienda['Producto'].value_counts().head(3)
  productos_menos_vendidos = tienda['Producto'].value_counts().tail(3)

  return productos_mas_vendidos, productos_menos_vendidos
```

```python
def visualizar_productos_mas_y_menos_vendidos(tienda, nombre_tienda):
  productos_mas_vendidos, productos_menos_vendidos = productos_mas_y_menos_vendidos(tienda)
  fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

  # Productos más vendidos
  ax1.bar(range(len(productos_mas_vendidos)), productos_mas_vendidos.values,
          color='green', alpha=0.7)
  ax1.set_title(f'Top 3 Productos Más Vendidos - {nombre_tienda}', fontsize=14, fontweight='bold')
  ax1.set_xlabel('Productos')
  ax1.set_ylabel('Cantidad Vendida')
  ax1.set_xticks(range(len(productos_mas_vendidos)))
  ax1.set_xticklabels(productos_mas_vendidos.index, rotation=45, ha='right')

  for i, v in enumerate(productos_mas_vendidos.values):
        ax1.text(i, v + 0.1, str(v), ha='center', va='bottom')


  # Productos menos vendidos
  ax2.bar(range(len(productos_menos_vendidos)), productos_menos_vendidos.values,
          color='red', alpha=0.7)
  ax2.set_title(f'Top 3 Productos Menos Vendidos - {nombre_tienda}', fontsize=14, fontweight='bold')
  ax2.set_xlabel('Productos')
  ax2.set_ylabel('Cantidad Vendida')
  ax2.set_xticks(range(len(productos_menos_vendidos)))
  ax2.set_xticklabels(productos_menos_vendidos.index, rotation=45, ha='right')

  for i, v in enumerate(productos_menos_vendidos.values):
        ax2.text(i, v + 0.05, str(v), ha='center', va='bottom')

  plt.show()
```

```python
visualizar_productos_mas_y_menos_vendidos(tienda, 'Tienda 1')
```

```python
visualizar_productos_mas_y_menos_vendidos(tienda2, 'Tienda 2')
```

```python
visualizar_productos_mas_y_menos_vendidos(tienda3, 'Tienda 3')
```

```python
visualizar_productos_mas_y_menos_vendidos(tienda4, 'Tienda 4')
```

# 5. Envío promedio por tienda

```python
costo_envio_tienda1 = tienda['Costo de envío'].mean()
costo_envio_tienda2 = tienda2['Costo de envío'].mean()
costo_envio_tienda3 = tienda3['Costo de envío'].mean()
costo_envio_tienda4 = tienda4['Costo de envío'].mean()
```

```python
print("COSTO DE ENVÍO PROMEDIO POR TIENDA \n")
print(f"Costo de Envío Promedio Tienda 1: ${costo_envio_tienda1:.0f}")
print(f"Costo de Envío Promedio Tienda 2: ${costo_envio_tienda2:.0f}")
print(f"Costo de Envío Promedio Tienda 3: ${costo_envio_tienda3:.0f}")
print(f"Costo de Envío Promedio Tienda 4: ${costo_envio_tienda4:.0f}")
```

```python
costo_envio_por_tienda = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Costo de Envío Promedio': [costo_envio_tienda1, costo_envio_tienda2, costo_envio_tienda3, costo_envio_tienda4]
})
costo_envio_ordenado = costo_envio_por_tienda.sort_values('Costo de Envío Promedio', ascending=True)
print("\nTiendas ordenadas por costo de envío (menor a mayor):")
print(costo_envio_ordenado)
```

```python
plt.figure(figsize=(10, 8))
plt.pie(costo_envio_ordenado['Costo de Envío Promedio'],
        labels=costo_envio_ordenado['Tienda'],
        autopct='%1.1f%%',
        startangle=90,
        colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99'])

plt.title('Distribución del Costo de Envío Promedio por Tienda',
          fontsize=14, fontweight='bold', pad=20)
plt.axis('equal')
plt.show()
```
